```python
import os
import cv2
import matplotlib.pyplot as plt
img_arr = []
game_plays =  []
views = []
steps = []
player_labels = []
contact = []
counter = 0
for name in os.listdir('train/frames/contact_0'):
#     print(name)
    img = cv2.imread(os.path.join('train/frames/contact_0' ,name),0)
    name = name.replace('.jpg','')
    l = name.split('_')
    game_plays.append('_'.join(l[:2]))
    views.append(l[2])
    steps.append(int(l[3]))
    player_labels.append(l[4])
    contact.append(0)
#     print(img.shape)
    img_resized = cv2.resize(img,(150,150))
    img_arr.append(img_resized.flatten())
#     plt.imshow(img_resized)
#     plt.show()
#     if counter == 0:
#         break
#     counter += 1
```

```python
for name in os.listdir('train/frames/contact_1'):
#     print(name)
    img = cv2.imread(os.path.join('train/frames/contact_1' ,name),0)
    name = name.replace('.jpg','')
    l = name.split('_')
    game_plays.append('_'.join(l[:2]))
    views.append(l[2])
    steps.append(int(l[3]))
    player_labels.append(l[4])
    contact.append(1)
#     print(img.shape)
    img_resized = cv2.resize(img,(150,150))
    img_arr.append(img_resized.flatten())
```

```python
len(img_arr)
```

```python
flat_data = np.array(img_arr)
contact = np.array(contact)
img_frame = pd.DataFrame(flat_data)
img_frame['contact'] = contact
```

```python
x = img_frame.iloc[:,:-1]
y = img_frame['contact']
```

```python
from sklearn import svm
# from sklearn.model_selection import GridSearchCV
# param_grid={'C':[0.1,1,10,100],'gamma':[0.0001,0.001,0.1,1],'kernel':['rbf','poly']}
model=svm.SVC(probability=True,max_iter = 100)
# model=GridSearchCV(svc,param_grid)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=77,stratify=y)
print('Splitted Successfully')
model.fit(x_train,y_train)
print('The Model is trained well with the given images')
# model.best_params_ contains the best parameters obtained from GridSearchCV
```

```python
from sklearn.metrics import accuracy_score
y_pred=model.predict(x_test)
print("The predicted Data is :")
print(y_pred)
print("The actual data is:")
print(np.array(y_test))
print(f"The model is {accuracy_score(y_pred,y_test)*100}% accurate")
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
